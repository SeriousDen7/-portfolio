## Модель

### Подготовка метрик и разделение на выборки

Добавим функцию подсчета по формуле

$$sMAPE =\frac{1}{N} \sum_{i=1}^n \frac{|y_{i}-y_{i}^{pred}|}{(|y_{i}|+|y_{i}^{pred}|)/2} $$

Итоговый показатель считаем по формуле ниже

$$General sMAPE = 0.25*sMAPE(rougher)+0.75*sMAPE(final)$$

````python
def sMAPE (target,preds):
    return ((np.abs(target - preds))/((np.abs(target) + np.abs(preds))/2)).mean()*100
````

````python
#Модель будем проверять через кросс-валидацию
def cross_val(model, features, target, cv=5):
    smape_score = make_scorer(sMAPE)
    scores = cross_val_score(model, features, target, cv=cv, scoring=smape_score)
    return scores.mean()
````

````python
feat_rougher_cols = []
for n in data_test.columns:
    if 'rougher' in n:
        feat_rougher_cols.append(n)
len(feat_rougher_cols)
````

````python
feat_final_cols = list(data_test.columns)
del feat_final_cols [0]
len(feat_final_cols)
````

````python
data_test = data_test.merge(data_full[['date','rougher.output.recovery','final.output.recovery']], how='inner', on='date')
data_test.info()
````

````python
features_train_rougher = data_train[feat_rougher_cols]
target_train_rougher = data_train['rougher.output.recovery']

features_train_final = data_train[feat_final_cols]
target_train_final = data_train['final.output.recovery']

features_test_rougher = data_test[feat_rougher_cols]
target_test_rougher = data_test['rougher.output.recovery']

features_test_final = data_test[feat_final_cols]
target_test_final = data_test['final.output.recovery']
````

````python
for df in [features_train_rougher,target_train_rougher,features_train_final,target_train_final,features_test_rougher,target_test_rougher, features_test_final,target_test_final]:
    print(df.shape)
````

### Обучение и кросс-валидация

#### Линейная регрессия

````python
LR_R = LinearRegression()
LR_F = LinearRegression()
LR_R.fit(features_train_rougher, target_train_rougher)
LR_F.fit(features_train_final, target_train_final)
smape_LR_rougher = cross_val(LR_R, features_train_rougher, target_train_rougher)
smape_LR_final = cross_val(LR_F, features_train_final, target_train_final)
f_smape_LR = 0.25*smape_LR_rougher+ 0.75*smape_LR_final
print('Итоговый sMAPE при линейной регрессии:', f_smape_LR)
````

#### Дерево регрессии

````python
for d in range(2,10):
    tree_R = DecisionTreeRegressor(max_depth = d)
    tree_F = DecisionTreeRegressor(max_depth = d)
    smape_tree_rougher = cross_val(tree_R, features_train_rougher, list(target_train_rougher))
    smape_tree_final = cross_val(tree_F, features_train_final, target_train_final)
    f_smape_tree = 0.25*smape_tree_rougher+ 0.75*smape_tree_final
    print(f'Итоговый sMAPE и отдельно по флотации и всему процессу при дереве регрессии с глубиной {d}:', f_smape_tree, smape_tree_rougher, smape_tree_final)
````

Для всего процесса лучше 2, а для флотации - 4

**Сохраним лучшие результаты для каждой модели и обучим их.**

````python
tree_R = DecisionTreeRegressor(max_depth = 4, min_samples_split = 2)
tree_F = DecisionTreeRegressor(max_depth = 2, min_samples_split = 2)
tree_R.fit(features_train_rougher, target_train_rougher)
tree_F.fit(features_train_final, target_train_final)
smape_tree_rougher = cross_val(tree_R, features_train_rougher, target_train_rougher)
smape_tree_final = cross_val(tree_F, features_train_final, target_train_final)
f_smape_tree = 0.25*smape_tree_rougher+ 0.75*smape_tree_final
print(f_smape_tree)
````

#### Dummy

````python
dummy_R = DummyRegressor()
dummy_F = DummyRegressor()
dummy_R.fit(features_train_rougher, target_train_rougher)
dummy_F.fit(features_train_final, target_train_final)
smape_dummy_rougher = cross_val(dummy_R, features_train_rougher, list(target_train_rougher))
smape_dummy_final = cross_val(dummy_F, features_train_final, target_train_final)
f_smape_dummy = 0.25*smape_dummy_rougher+ 0.75*smape_dummy_final
print('Итоговый sMAPE в дамми модели:', f_smape_dummy)
````

Остановимся на модели дерева регрессии. Линейная регрессия показывает результат хуже дамми модели.
