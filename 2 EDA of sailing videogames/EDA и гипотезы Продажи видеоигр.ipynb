# Исследование продаж видеоигр

Видеоигры стали неотъемлемой частью жизни многих людей, где то в ни играют больше, где то меньше, ктото в ни играет для расслабления, а ктото для концентрации или чтобы проверить себя. Некоторые игры переросли в киберспортивные дисциплины с многомилионными выигрышами. 
А некоторые вовсе считаю, что игры повышают уровень агрессивности. Но доказательства ученых, что видеоигры полезны, давно развеяли миф о том, что их стоит сторониться. В любом случае, во время игры игрок получает выброс дофаминов(например при выполнении заданий/квестов или когда побеждает сложного босса).
Представляю Вам свое исследование посвященое статистике продаж видеоигр по всему миру до 2016 года.

**Цель исследования:** 
1. Выявить особенности и зависимости продаж видеоигр по странам, жанрам и возрастам.



**Задачи исследования:**
1. Оценить качество исходных данных
2. Подготовить данные к исследованию
3. Исследовать данные на наличие особенностей и взаимосвязей
4. Сформировать портрет среднестатистического покупателя игр по регионам
5. Сформулировать и проверить некоторые гипотезы

**Объектом исследования** является статистика продаж видеоигр по всему миру.

**Предметом исследования** является комплекс методик для обработки и анализа отдельных данных и их взаимосвязей

**Ход выполнения**
1. Загрузить данные из файла '/datasets/games.csv'
2. Проанализировать общую информацию о качестве и структуре данных
3. Выполнить предобработку данных, которая включает в себя: обработку пропусков, дубликатов, аномальных значений и выбросов. преобразование информации в удобный для исследования формат
4. Исследовать отдельные факторы их взаимосвязь и изучить их влияние друг на друга
5. Для каждого региона составить портрет игрока: выявить топ-5 платформ и жанров
6. Проверить гипотезы относительно рейтингов отдельных платформ и жанров
5. Сделать выводы и рекомендации для дальнейшего исследования

**Описание данных**
- `Name` — название игры;
- `Platform` — платформа;
- `Year_of_Release` — год выпуска;
- `Genre` — жанр игры;
- `NA_sales` — продажи в Северной Америке (миллионы проданных копий);
- `EU_sales` — продажи в Европе (миллионы проданных копий);
- `JP_sales` — продажи в Японии (миллионы проданных копий);
- `Other_sales` — продажи в других странах (миллионы проданных копий);
- `Critic_Score` — оценка критиков (максимум 100);
- `User_Score` — оценка пользователей (максимум 10);
- `Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

## Предобработка данных

Обзор данных позволил вскрыть различные дефекты исходных данных: названия колонок, типы данных, пропуски(в некоторых колонках большое количество), обязательно будет проверка на дубликаты и аномальные значения. Со всем этим мы разберемся в предобработке данных.

### Изменения названий колонок

```python
data.columns
```

Привевдем к змеиному регистру названия колонок, кроме названия регионов, они так и останутся заглавными буквами

```python
data = data.rename(columns={'Name':'name',
                     'Platform':'platform',
                     'Year_of_Release':'year_of_release', 
                     'Genre':'genre', 
                     'Other_sales':'other_sales', 
                     'Critic_Score':'critic_score', 
                     'User_Score':'user_score', 
                     'Rating':'rating'}
           )
```

###  Изменение типов данных колонок

```python
data['user_score'].unique()
```

```python
data.loc[data['user_score']=='tbd', 'user_score'] = np.nan
```

```python
data['user_score'] = data['user_score'].astype('float64', errors='ignore')
```

```python
data['user_score'] = data['user_score']*10
```

```python
for col in ['year_of_release', 'user_score', 'critic_score']:
    data[col] = data[col].astype('Int64', errors='ignore')
```

Тип данных в оценке пользователей не изменялся, проблема таилась в ошибке при заполнении: вместо числового было "tbd" - заменил на пропуск, и потом перевел в соответствующий тип данных: год в целочисленный, оценку игроков умножил на 10(100 бальная шкала, как в оценке критиков) и обе оценки перевел в целочисленный.

tbd - "Подлежит уточнению", значит объективной оценки еще нет, и такое значение лучше убрать, лучше всего посчитал изменить на пропуск, у нас и так они имеются в этой колонке

**Вывод:** Исправил названия колонок и тип данны в трех колонках

### Обработка пропусков

Посмотрим сколько проусков в процентах и в абсолютных значения

```python
pd.DataFrame(round(data.isna().mean()*100)).style.background_gradient('coolwarm')
```

```python
data.isna().sum()
```

Совсем немного пропусков в годе выпуска и большое количество пропусков в оценках и возрастному рейтингу.
Так же видим 2 пропуска в названии и жанре, от этих дву строк мы избавимся. Удалим и строки с пропуском года релиза.

```python
data = data.dropna(subset = ['name'])
```

```python
data = data.dropna(subset=['year_of_release'])
```

Работаем с возрастным рейтингом игр

```python
data['rating'].sort_values().unique()
```

```python
data['rating'] =data['rating'].where(data['rating'] != 'K-A', 'E')
```

Проверили информацию по рейтингу: с 1998 года рейтинг K-A переименовали в E

```python
#делаем залушку на пропусках в рейтинге
data['rating'] = data['rating'].fillna('not_rait')
```

### Проверка дубликатов и аномальных значений

```python
data.duplicated().sum()
```

```python
data['genre'].sort_values().unique()
```

```python
len(data['name'].sort_values().unique())
```

Явных дубликатов нет, неявных предположем что тоже, более 11 тысяч разных названий просматривать и приводить к одному(если действительно одинаковые игры) - будет бессмысленно, и большая вероятность ошибиться.

```python
data.loc[((data['year_of_release'] == 1985) & (data['platform'] == "DS")), 'year_of_release'] = 2004
```

Аномальному значению присвоил адекватное, год выпуска поменял на год выпуска платформы, продажи маленькие, поэтому не имеет значения в какой год (из годов, когда приставка существовала) ее перенести.

```python
#Добавляем столбец с суммарными продажами
data.insert(8, 'total_sales', data['NA_sales']+data['EU_sales']+data['JP_sales']+data['other_sales'])
```

**Вывод по разделу:** Изменили названия столбцов, их тип данных, избавились от незначительны пропусков, проверили наличие дубликатов, и в процессе предобработки и обзора данных выбросов не обнаружено(кроме неправильно введенного года у одной игры). Добавил столбец с суммой всех продаж.

### Просмотр объема продаж игр по платформам

Проверим как менялись продажи по платформам. Выберем самые популярные платормы и посмотрим их "срок жизни"

```python
top_platform = data.groupby('platform')['total_sales'].agg('sum').sort_values(ascending=False)[:6]
top_platform
#"PS2","X360","Wii","PS3","DS","PS"
```

Сгрупировал данные платформам, посчитал сумму проданных копий в мире, отсортировал и выбрал 6 самых популярных за все время платформ: "PS2","X360","Wii","PS3","DS","PS"

```python
data_PS2 = data.query('platform =="PS2"')
```

```python
(data_PS2.pivot_table(index='year_of_release',values='total_sales',aggfunc='sum')
 .plot(style='o-', grid=True, title = "Количество проданных копий по годам на PS2(млн.)"));
```

Тестово решил посмотреть выпуск игр на PS2 по годам. Видим, что пик популярности (количество проданных копий игр) был в 2001-2005 годах. А потом постепенно сошло на нет, ввиду выпуска консоли нового поколения.

```python
plt.rcParams ['figure.figsize'] = [15, 6]
for platform in ["PS2","X360","Wii","PS3","DS","PS",'PS4','XOne']:
    data_for_plt = (data.query('platform ==@platform')
        .pivot_table(index='year_of_release',values='total_sales',aggfunc='sum'))
    plt.plot(data_for_plt)
plt.legend(["PS2","X360","Wii","PS3","DS","PS",'PS4','XOne'])
plt.title('Количество проданных копий по годам на популярных платормах(млн.)')
plt.minorticks_on()
plt.grid(which='both', linestyle = ':')
plt.show()
```

Рассмотрим платформы отдельно.

```python
for platform in ["PS2","X360","Wii","PS3","DS","PS"]:
    (data.query('platform ==@platform')
    .pivot_table(index='year_of_release',values='total_sales',aggfunc='sum')
    .plot(style='o-', grid=True,title=platform, figsize= (15,5)));
```

**Вывод:** В среднем выбранные платформы (самые большие суммы продаж за все время) популярны в течении 10 лет. Видно что каждая из них была популярна в какой то определенный год, т.е пик в разное время, но в 2008-2011 гг. главенствовало 3 платформы: PS3, X360 и Wii.

## Проверка гипотез    

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

```python
from scipy import stats as st
```

H_0: Средний пользовательский рейтинг PC = среднему пользовательскому рейтингу XOne

H_a: Средний пользовательский рейтинг PC ≠ среднему пользовательскому рейтингу XOne

alpha = 0.05

```python
results = st.ttest_ind( list(data_act.query('user_score>=0 & platform == "PC"')['user_score'])
                       , list(data_act.query('user_score>=0 & platform == "XOne"')['user_score'])
                       , equal_var=False)
alpha = 0.05

print (results.pvalue)
if results.pvalue < alpha: 
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Не можем отвернуть гипотезу, что рейтинги PC и XOne равны

### Средние пользовательские рейтинги жанров Action и Sports разные.

Приступим к проверке следующих гипотез

H_0: Средний пользовательский рейтинг жанра Action = среднему пользовательскому рейтингу жанру Sports

H_a: Средний пользовательский рейтинг жанра Action ≠ среднему пользовательскому рейтингу жанру Action

alpha = 0.05

```python
results = st.ttest_ind(list(data_act.dropna(subset = ['user_score']).query('genre == "Action"')['user_score'])
                       , list(data_act.dropna(subset = ['user_score']).query('genre == "Sports"')['user_score'])
                       , equal_var=False)
alpha = 0.05

print (results.pvalue)
if results.pvalue < alpha: 
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Заначит принимаем альтернативную гипотезу, рейтинги сравниваемых жанров не равны в среднем
