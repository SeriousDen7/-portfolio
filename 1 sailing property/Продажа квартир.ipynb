### Предобработка данных

#### Обработка пропусков

Узнаем % пропусков к каждой колонке

```python
pd.DataFrame(round(data.isna().mean()*100)).style.background_gradient('coolwarm')
```

Будем предполагать, что не указанное количество балконов - 0, и точно так же с аппартаментами.

```python
#Заполняем пропуски в графах "балконы" и "аппартаменты"
data['balcony'] = data.fillna(0)
data['is_apartment'] = data.fillna(False)
```

```python
#удаляем строки с пропусками в колонках "этажность здания" и "населенный пункт"
data = data.dropna(subset = ['floors_total', 'locality_name'])
```

**Вывод:** Заменил пропуски в "Аппартаментах" на False, а в балконах на 0. 

Если говорить про пропуски в добавленных пользователем данных, то скорее всего люди просто не знают корректные данные: сколько этажей, жилую и площадь кухни. 
А название населенного пункта просто не добавили, не заметили (хотя это может быть и жители СПБ не сочли необходимым указывать СПБ как населенный пункт)

Принял решение **убрать строки** с пропущенными значениями в столбцах, которые составляют менее 1% (количество этажей и населенный пункт).


Остальные пропуски не трогаем, т.к. однозначно не известен характер и возникновения

#### Изменение типов данных

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

```python
# Изменим тип данных в некоторых колонках на целочисленный
for column in ['floors_total', 'balcony', 'parks_around3000', 'ponds_around3000', 'days_exposition']:
    data[column] = data[column].astype('Int64', errors='ignore')
```

```python
data['is_apartment'] = data['is_apartment'].astype('bool', errors='ignore')
```

**Вывод:** Исправил тип данны даты публикации для дальнейшего удобства преобразования.
В 5 колонках поменял тип данных на целочисленный для легкости восприятия и снижения занимаемой памяти.
И тип данных Аппартаментов изменил на булев.

### Расчет дополнительной информации по объектам

Для дальнейшего анализа посчитаем некоторую дополнительную информацию по объектам.

```python
# Цена за кв.м.
data['price_per_m2'] = data['last_price']/data['total_area']
```

```python
# день недели, когда выложили объявление
data['weekday'] = pd.DatetimeIndex(data['first_day_exposition']).weekday
```

```python
# месяц в году(от 1 до 12) и месяц + год, когда выложили объявление
data['month'] = pd.DatetimeIndex(data['first_day_exposition']).month
data['month_year'] = data['first_day_exposition'].astype('datetime64[M]')
```

```python
# год, когда выложили объявление
data['year'] = pd.DatetimeIndex(data['first_day_exposition']).year
```

```python
# функция, которая категорезирует этаж на первый последний и другой в нову колонку
def floor_typing(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor ==1:
        return 'первый'
    if floor == floors_total:
        return 'последний'
    return 'другой'
```

```python
# Применяем функцию выше
data['floor_type'] = data.apply(floor_typing, axis = 1)
```

```python
# Расстояние до центра в километрах
data['cityCenters_nearest_km'] = data['cityCenters_nearest']//1000
```

Все что неободимо было - добавил
